```python
## Q9. Write a python program to scrape mentioned details from dineout.co.in :#
 #i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL ## 
```

```python
dineout_page_raw=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
dineout_page_content=BeautifulSoup(dineout_page_raw.content)
Rest_name=dineout_page_content.findAll('a',class_="restnt-name ellipsis")
sub_place_data=dineout_page_content.findAll('a',{'data-type':"AreaClick"})
place_data=dineout_page_content.findAll('a',{'data-type':"LocalityClick"})
cuisine_data=dineout_page_content.findAll('div',class_="detail-info")
rating_data=dineout_page_content.findAll('div',class_="restnt-rating rating-4")
img_url_data=dineout_page_content.findAll('img',class_="no-img")
```

```python
def attach(x,y):
    for i in x:
        y.append(i.text)

Restaurant_names=[]
sub_places=[]
sub_city=[]
Restaurant_place=[]
all_ratings=[]
all_cuisine_data=[]
image=[]


attach(Rest_name,Restaurant_names)
attach(sub_place_data,sub_city)
attach(place_data,sub_places)
attach(rating_data,all_ratings)

z=len(Restaurant_names)
for i in range(0,z):
    Restaurant_place.append((sub_city[i]+','+sub_places[i]))
```

```python
for i in cuisine_data:
    all_cuisine_data.append(i.li.text)

for i in img_url_data:
    image.append(i['data-src'])
p=len(image)
```

```python
for i in range(0,p):
    print("Name of the Restaurant:",Restaurant_names[i],"\n","cuisine:",all_cuisine_data[i],"\n",'Location:',Restaurant_place[i],"\n","rating:",all_ratings[i],"\n","image url:",image[i],"\n\n\n")
```

```python
framing=pd.DataFrame(list(zip(Restaurant_names,all_cuisine_data,Restaurant_place,all_ratings,image)))
framing.columns=['Restaurant Name', 'cuisine', 'Location', 'location','image url']
framing.index.name='Sr.No'
framing.index+=1
framing
```

```python
##Q.10 Write a python program to scrape first 10 product details which include product name , price , Image URL from #
 #https://www.bewakoof.com/women-tshirts?ga_q=tshirts . ##
```

```python
BK_W_page_raw=requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts')
BK_W_page_content=BeautifulSoup(BK_W_page_raw.content)
product_name_data=BK_W_page_content.findAll('div',class_='productCardDetail') 
price_data=BK_W_page_content.findAll('span',class_='discountedPriceText')
image_url_data=BK_W_page_content.findAll('img',class_='productImgTag')
```

```python
imageurl=[]
for i in image_url_data:
    imageurl.append(i['src'])

productname=[]
for i in product_name_data:
    productname.append(i.h3.text)
    
price_of_products=[]
for i in price_data:
    price_of_products.append(i.b.text)
s=len(price_of_products)
```

```python
for i in range(0,s): 
    print("Name:",productname[i],"\n","Price:",price_of_products[i],"\n","Image URL:",imageurl[i],"\n\n\n")
```

```python
productsheet=pd.DataFrame(list(zip(productname,price_of_products,imageurl)))
productsheet.columns=['Product Name', 'Price', 'image url']
productsheet.index+=1
productsheet
```
