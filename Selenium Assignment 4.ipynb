1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# calling chrome driver
driver=webdriver.Chrome('chromedriver.exe')
# getting the url
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
# defining empty list
rank_list=[]
name_list=[]
artist_list=[]
upload_list=[]
views_list=[]
time.sleep(5)
#scraping required using for loop
for i in range(1,31):
    rank=driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[1]')
    rank_list.append(rank.text)
    name=driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[2]')
    name_list.append(name.text)
    artist=driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[3]')
    artist_list.append(artist.text)
    views=driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[4]')
    views_list.append(views.text)
    upload=driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[5]')
    upload_list.append(upload.text)
#closing the driver
driver.close()
#extracting only name the name_list
names=[]
for k in name_list:
    n=k.split('"')
    for r in n:
        names.append(r)
x=names[1:90:3]
#framing the data
print('\033[1m'+"-----------------------------------------------------------------------------------------------------------\n                                   Most viewed videos on YouTube\n-----------------------------------------------------------------------------------------------------------"+'\033[1m')
df=pd.DataFrame(list(zip(rank_list,x,artist_list,views_list,upload_list)))
df.columns=['Ranking','Name of the video','Artist Name','Views in billions','Upload Date']
df.index+=1
df
```

5. Scrape the details of trending repositories on Github.com.

You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# calling chrome driver
driver=webdriver.Chrome('chromedriver.exe')
# getting the url
url='https://github.com/'
driver.get(url)
time.sleep(3)
#clicking on trending via explore 
explore=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary').click()
action=ActionChains(driver)
trend=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a')
action.move_to_element(trend).click(trend).perform()
time.sleep(3)
#fetching urls of each trending repository
links=driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a')
links_list=[]
def attach(x,y):
    for i in x:
        y.append(i.get_attribute('href'))
attach(links,links_list)        
```

```python
#fetching the required data from each link of the trending repository
title_list=[]
desp_list=[]
c_count_list=[]
all_langs=[]
for k in links_list:
    driver.get(k)
    time.sleep(5)
    title=driver.find_element_by_xpath('//*[@id="repository-container-header"]/div[1]/div/h1/span[1]/a')
    title_list.append(title.text)
    try:
        description=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[3]/div[2]/div/div[1]/div/p')
        desp_list.append(description.text)
    except:
        desp_list.append('None')
    try:
        c_count=driver.find_element_by_xpath('//h2[@class="h4 mb-3"]/a[contains(text(),"Contributors")]/span')
        c_count_list.append(c_count.text)
    except:
        c_count_list.append('None')
    try:
        x=driver.find_element_by_xpath("//li[@class='d-inline']//a//span[1]").text
        z=[]
        y=x.split('\n')
        for i in y:
            z.append(i)
        w=[]
        for i in range(0,len(z),2):
            w.append(z[i])
        all_langs.append(w)
    except:
        all_langs.append('None')
#framing the data
github_data=pd.DataFrame(list(zip(title_list,desp_list,c_count_list,all_langs)))
github_data.columns=['Repository Name','Description','No of Contributors','Languages Used']
github_data.index+=1
github_data
```
