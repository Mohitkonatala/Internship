```python
# installing required libraries
import selenium 
from selenium import webdriver
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings("ignore")

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.common import exceptions 
from selenium.common.exceptions import StaleElementReferenceException
import time
```

```python
# calling chrome driver
driver=webdriver.Chrome('chromedriver.exe')
```

```python
# getting the url
url='https://www.naukri.com'
driver.get(url)
```

```python
#locating the required search bars
search_job= driver.find_element_by_id("qsb-keyword-sugg")
search_location=driver.find_element_by_id("qsb-location-sugg")
```

```python
# sending keys to search bar
search_job.send_keys("Data Analyst")
search_location.send_keys("Bangalore")
```

```python
#clicking on the search
search_button=driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[2]/section/div/form/div[3]/button')
search_button.click()
```

```python
#closing the bot chatbox
cancel= WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[5]/div/div[1]/div[1]/div'))).click()
```

```python
# defining appending function and empty list
job_title_list=[]
job_location_list=[]
company_name_list=[]
exp_list=[]
def attach(x,y):
    for i in x:
        y.append(i.text)
```

```python
#scraping the required Data
job_titles= driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
attach(job_titles,job_title_list)

company_names=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
attach(company_names,company_name_list)

job_locations=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
attach(job_locations,job_location_list)

job_exps=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')
attach(job_exps,exp_list)
```

```python
#picking the top 10 Data
top_10_jobs=job_title_list[0:10]
top_10_locs=job_location_list[0:10]
top_10_exp=exp_list[0:10]
top_10_com=company_name_list[0:10]
```

```python
# Framing the required Data
import pandas as pd

df=pd.DataFrame(list(zip(top_10_jobs,top_10_locs,top_10_com,top_10_exp)))
df.columns=['Job Title','Location','Company','Experience Required']
df.index+=1
df
```

```python
#calling the driver
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-%20earpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKC%20TSVZAXUHGREPBFGI&marketplace.'
driver.get(url)
```

```python
#opening all reviews
all_review=driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]').click()
```

```python
#creating lists and defining function for appending the data
rating_list=[]
summary_list=[]
review_list=[]
def attach(x,y):
    for w in x:
        y.append(w.text)
#creating page numbers        
pages=np.arange(1,11,1)
```

```python
#creating loop for scraping the required data
start_string='&page='
new_url=driver.current_url
for page in pages:
    try:
        driver.get(new_url+start_string+str(page))
    except:
        break
    time.sleep(5)
    try:
        ratings=driver.find_elements_by_xpath('//div[starts-with(@class,"_3LWZlK")]')
        attach(ratings,rating_list)
        summary=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
        attach(summary,summary_list)
        review=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
        attach(review,review_list)
    except exceptions.StaleElementReferenceException:  
        pass
```

```python
#Framing the acquired data
del rating_list[0:110:11]
df=pd.DataFrame(list(zip(rating_list,summary_list,review_list)))
df.columns=['Rating','Review Summary','Full Review']
df.index+=1
pd.set_option('display.max_rows', None)
df
```

```python
# calling the url using driver
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.ambitionbox.com/'
driver.get(url)
driver.maximize_window()
```

```python
# clicking on salaries button
salaries_button=driver.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]').click()
```

```python
# search for data scientist in job profiles
search_bar=driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input')
search_bar.send_keys('Data Scientist')
```

```python
#clicking the first suggestion
search_bar.send_keys(Keys.ARROW_DOWN, Keys.ENTER)
```

```python
#creating companies name list
company_names=[]
for i in range(1,11):
    str1='/html/body/div/div/div/main/section[1]/div[2]/div[3]/div[2]/div['
    str2=str(i)
    str3=']/div[1]/div/div/div[1]/a'
    xpath=str1+str2+str3
    company_name=driver.find_element_by_xpath(xpath)
    company_names.append(company_name.text)
```

```python
#defining attach function for appending the acquired name
def attach(x,y):
    for i in x:
        y.append(i.text)
```

```python
#defining experience required list
exp_req=driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
exp_list=[]
attach(exp_req, exp_list)
new_exp_list=[]
for i in range(0,10):
    x=exp_list[i]
    new_exp_list.append(x[19:-4])
```

```python
# creating average salary list
avg_sal=driver.find_elements_by_xpath('//div[@class="average-indicator-wrapper"]')
avg_sal_list=[]
attach(avg_sal, avg_sal_list)
```

```python
# creating minimum and maximum salary lists
sal_val=driver.find_elements_by_xpath('//div[@class="salary-values"]')
sal_val_list=[]
attach(sal_val, sal_val_list)
min_sal_list=[]
max_sal_list=[]
for i in range(0,10):
    s=sal_val_list[i]
    min_sal_list.append(s[0:-8])
    max_sal_list.append(s[-7:])
```

```python
# creating salaries record list
sal_record_list=[]
for i in range(1,11):
    str1='/html/body/div/div/div/main/section[1]/div[2]/div[3]/div[2]/div['
    str2=str(i)
    str3=']/div[1]/div/div/div[1]/span'
    xpath=str1+str2+str3
    sal_record=driver.find_element_by_xpath(xpath)
    sal_record_list.append(sal_record.text)
```

```python
# framing the acquired data
df=pd.DataFrame(list(zip(company_names, sal_record_list, avg_sal_list, min_sal_list, max_sal_list, new_exp_list)))
df.columns=['Company Name', 'Total Salary Record', 'Average Salary', 'Mininum Salary', 'Maximum Salary', 'Experience Required']
df.index+=1
df
```
